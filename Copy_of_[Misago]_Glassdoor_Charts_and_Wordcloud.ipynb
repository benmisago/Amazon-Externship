```python
from collections import Counter
import pandas as pd
from google.colab import files

uploaded = files.upload()


df = pd.read_csv("glassdoor_cleaned_text (2).csv")


all_words = " ".join(df["review_pros_cleaned"].dropna().astype(str)).split()

word_counts = Counter(all_words)


print(word_counts.most_common(20))
```

```python
import pandas as pd

word_df = pd.DataFrame(word_counts.items(), columns=['word', 'count']).sort_values(by='count', ascending=False)
```

```python
import matplotlib.pyplot as plt

most_common = word_counts.most_common(10)
words, counts = zip(*most_common)

plt.bar(words, counts)
plt.title("Top 10 Keywords (Unigrams)")
plt.xticks(rotation=45)
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
from nltk.util import ngrams

bigrams = list(ngrams(all_words, 2))
trigrams = list(ngrams(all_words, 3))

bigram_counts = Counter(bigrams)
trigram_counts = Counter(trigrams)
```

```python

common_bigrams = bigram_counts.most_common(10)
bg_labels = [' '.join(bg) for bg, count in common_bigrams]
bg_counts = [count for bg, count in common_bigrams]

plt.bar(bg_labels, bg_counts)
plt.title("Top 10 Bigrams")
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import random


mask = np.array(Image.open("star.jpg"))


def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return random.choice(["pink", "purple", "blue", "yellow"])

wordcloud = WordCloud(
    width=800,
    height=800,
    background_color="white",
    mask=mask,
    contour_width=1,
    contour_color="blue",
    collocations=False,

).generate(' '.join(all_words))


wordcloud.recolor(color_func=custom_color_func)


plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Star-Shaped Word Cloud", fontsize=20)
plt.show()
```
